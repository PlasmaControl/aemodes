```python
import pickle
import matplotlib.pyplot as plt
```

```python
file_path = '/scratch/gpfs/nc1514/aemodes/data/co2_250_detector.pkl'

channels = ('r0', 'v1', 'v2','v3')
```

```python
data = pickle.load(open(file_path, 'rb'))
```

```python
mode, idx = 0, 0

shotn = data[0][idx]
inp = data[mode+1][idx]
out = data[mode+2][idx]
```

```python
fig, axes = plt.subplots(5, 1, figsize=(12, 15))

# Plot spectrograms for each channel
for i, channel in enumerate(channels):
    axes[i].imshow(inp[channel].T, aspect='auto', origin='lower', cmap='viridis')
    axes[i].set_title(f'Channel: {channel}')
    axes[i].set_ylabel('Frequency')
    if i < len(channels) - 1:
        axes[i].set_xticks([])

# Plot binary label at the bottom
axes[4].imshow(out.T, aspect='auto', origin='lower', cmap='binary', interpolation='none')
axes[4].set_title('Binary Label (out)')
axes[4].set_xlabel('Time')
axes[4].set_ylabel('Label')

plt.tight_layout()
plt.show()
```
